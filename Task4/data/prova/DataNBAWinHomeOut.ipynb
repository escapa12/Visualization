```python
import pandas as pd
import numpy as np

df = pd.read_csv("DataNBA1970.csv")
df
```

```python
clean_df = df[["year","homeWon","awayWon", "franchID"]]\
    .groupby(["franchID", "year"])[["homeWon","awayWon"]].sum()
clean_df
```

```python
data_dict = clean_df.to_dict()
data_dict
```

```python
flat_data = {}

for revenue_type, revenue_dictionary in data_dict.items():
    for country_year, revenue in revenue_dictionary.items():
        country = country_year[0]
        year = country_year[1]
        plot_name = "_".join([country,revenue_type])
        if  plot_name not in flat_data:
            flat_data[plot_name] = {"name": plot_name, "values": []}
        flat_data[plot_name]["values"].append({"x": year, "y": revenue})   

d3_plot_list = []
for kaye,value in flat_data.items():
    value['values'].sort(key=lambda v: v['x'])
    d3_plot_list.append(value)
d3_plot_list
```

```python
import json
with open('home_away.json', 'w') as outfile:
    json.dump(d3_plot_list, outfile)
```
