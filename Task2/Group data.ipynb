#  Countries - Year benefit

```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
```

```python
data_in= pd.read_csv('retail_sales_clean.csv')
data_in.head()
```

```python
countries = pd.unique(data_in['Retailer country'])
years = pd.unique(data_in['Year'])
df = pd.DataFrame(index=np.hstack((countries,'World')), columns=years)

for year in years:
    df_any=[]
    for country in countries:
        df_any.append(data_in[(data_in['Retailer country']==country) &( data_in['Year']==year) ]['Gross profit'].sum())
    df_any.append(sum(df_any))   
    df[year]=df_any

pd.DataFrame.to_csv(df,'profit_country_year.csv')
df
```

#  Countries - product benefit

```python
countries = pd.unique(data_in['Retailer country'])
products = pd.unique(data_in['Product line'])

df = pd.DataFrame(index=np.hstack((countries,'World')), columns=products)

for product in products:
    df_any=[]
    for country in countries:
        df_any.append(data_in[(data_in['Retailer country']==country) &( data_in['Product line']==product) ]['Gross profit'].sum())

    df_any.append(sum(df_any))   
    df[product]=df_any
pd.DataFrame.to_csv(df,'profit_country_productLine.csv')
df
```

# 2007 order type- country

```python
countries = pd.unique(data_in['Retailer country'])
orders = pd.unique(data_in['Order method type'])

df = pd.DataFrame(index=np.hstack((countries,'World')), columns=orders)

for order in orders:
    df_quant=[]
    for country in countries:
#         df_benef.append(data_in[(data_in['Retailer country']==country) &( data_in['Order method type']==order) ]['Gross profit'].sum())
        df_quant.append(data_in[(data_in['Retailer country']==country) &( data_in['Order method type']==order) ]['Quantity'].sum())
    
    df_quant.append(sum(df_quant))    
    df[order]=df_quant
pd.DataFrame.to_csv(df,'order_country_year.csv')
df
```
