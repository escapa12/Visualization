```python
import pandas as pd
import numpy as np

df = pd.read_csv("retail_sales_clean.csv")
df["Planned revenue"] = df["Planned revenue"].apply(lambda x: x/1000000)
df["Revenue"] = df["Revenue"].apply(lambda x: x/1000000)
df["Planned revenue"].sum()
df["Revenue"].sum()
```

```python
world_df = df.groupby(["Year"])["Planned revenue","Revenue"].sum()
world_df["Retailer country"] = "World"
world_df['Year'] = world_df.index
df = pd.concat([df,world_df])
world_df
```

```python
clean_df = df[["Year","Revenue","Planned revenue", "Retailer country"]]\
    .groupby(["Retailer country", "Year"])[["Revenue","Planned revenue"]].sum()
clean_df
```

```python
data_dict = clean_df.to_dict()
data_dict
```

```python
flat_data = {}

for revenue_type, revenue_dictionary in data_dict.items():
    for country_year, revenue in revenue_dictionary.items():
        country = country_year[0]
        year = country_year[1]
        plot_name = "_".join([country,revenue_type])
        if  plot_name not in flat_data:
            flat_data[plot_name] = {"name": plot_name, "values": []}
        flat_data[plot_name]["values"].append({"x": str(year), "y": revenue})   

d3_plot_list = []
for kaye,value in flat_data.items():
    d3_plot_list.append(value)
d3_plot_list
```

```python
import json
with open('data/revenue.json', 'w') as outfile:
    json.dump(d3_plot_list, outfile)
```
