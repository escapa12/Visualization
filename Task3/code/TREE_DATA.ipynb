```python
import pandas as pd
import json
```

```python
df = pd.read_csv("data/retail_sales_clean.csv")
```

```python
def create_node (name,quant,profit,children):
    """ Return a dictionary
        name: name of the node
        size: Quantity sold
        color: gross profit
        children: list with the childrens (each children is a dictionary)"""
    dictionary={'name':str(name),'Quantity_sold':quant,'Profit':profit,'children':children}
    return(dictionary)
```

```python
# all_country=[]
# for country in pd.unique(df['Retailer country']):
#     df1=df[df['Retailer country']==country]
#     child_country=[]
#     for y in pd.unique(df1['Year']):
#         df2=df1[df1['Year']==y]
df2=df[(df['Year']==2007) &(df['Retailer country']=='Spain')]                    
        
#         year=str(y)
child_year=[]
for pr_line in pd.unique(df2['Product line']):
    df3=df2[df2['Product line']==pr_line]
    child_pr_line=[]
    for pr_type in pd.unique(df3['Product type']):
        df4=df3[df3['Product type']==pr_type]
        child_pr_type=[]
        for order in pd.unique(df4['Order method type']):
            df5=df4[df4['Order method type']==order]
            child_order=[]
            for product in pd.unique(df5['Product']):
                df6=df5[df5['Product']==product]
                child_product=[]
                for item in ['Revenue','Product cost','Quantity','Unit cost','Unit price','Gross profit']:
                    dic_product={}
                    dic_product['name']=item+': '+str(float(df6[item]))
                    dic_product['Quantity_sold']=1
                    dic_product['Profit']=0
                    dic_product['value']=
                    child_product.append(dic_product)
                child_order.append(create_node(product,df6['Quantity'].sum(),df6['Gross profit'].sum(),child_product))
            child_pr_type.append(create_node(order,df5['Quantity'].sum(),df5['Gross profit'].sum(),child_order))
        child_pr_line.append(create_node(pr_type,df4['Quantity'].sum(),df4['Gross profit'].sum(),child_pr_type))
    child_year.append(create_node(pr_line,df3['Quantity'].sum(),df3['Gross profit'].sum(),child_pr_line))
#         child_country.append(create_node(year,df2['Quantity'].sum(),df2['Gross profit'].sum(),child_year))
#     all_country.append(create_node(country,df1['Quantity'].sum(),df1['Gross profit'].sum(),child_country))
root=create_node('root',666,666,child_year)
with open('data/json_data_Spain2007.json','w') as file:
    file.write(json.dumps(root))                
# df5=df[(df['Year']==2007) & (df['Order method type']=='Telephone')&(df['Retailer country']=='Canada')&(df['Product type']=='Cooking Gear')]                    
```
