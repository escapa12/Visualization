```python
#create an array for the total number of votes Democrat,Republican
res_fips=pd.unique(res_data['state'])

biparty=np.zeros([len(res_fips),2])
i=0
for state in res_fips:
    st_info=res_data[res_data['state']==state]
#     print(np.sum(st_info[st_info['party']=='Democrat']['votes']))
    biparty[i,0]=np.sum(st_info[st_info['party']=='Democrat']['votes'])
    biparty[i,1]=np.sum(st_info[st_info['party']=='Republican']['votes'])
    i+=1
```

```python
biparty_percent=biparty/biparty.sum(axis=1)[:,np.newaxis]
```

```python
d_frame=pd.DataFrame(biparty_percent)
d_frame.index=res_fips
pd.DataFrame.to_csv(d_frame,'results.csv',header=['Democrat','Republican'])
```
